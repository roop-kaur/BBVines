```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

```python
red_wine_original = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv',sep=';')
white_wine_original = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',sep=';')
red_wine = red_wine_original.copy()
white_wine = white_wine_original.copy()
```

```python
red_wine.hist(bins=15, color='red', edgecolor='black', linewidth=1.0, xlabelsize=8, ylabelsize=8, grid=False)

plt.tight_layout(rect=(0, 0, 1.2, 1.2))

rt = plt.suptitle('Red Wine Univariate Plots', x=0.65, y=1.25, fontsize=15)
```

```python
white_wine.hist(bins=15, color='grey', edgecolor='black', linewidth=1.0, xlabelsize=8, ylabelsize=8, grid=False)

plt.tight_layout(rect=(0, 0, 1.2, 1.2))

rt = plt.suptitle('White Wine Univariate Plots', x=0.65, y=1.25, fontsize=15)
```

```python
sns.histplot(data = red_wine, x = 'residual sugar', color = 'red',bins=50)
```

```python
sns.set(style="whitegrid")  
fig,axs = plt.subplots(2,2, figsize = (10,8))
sns.histplot(data=red_wine, x="residual sugar", bins = 10, color="red", edgecolor='black',ax=axs[0, 0]).set_ylim(0,2500)
sns.histplot(data=white_wine, x="residual sugar", bins = 30,color="grey",edgecolor='black', ax=axs[0,1]).set_ylim(0,2500)
fig.tight_layout()
plt.suptitle('Residual sugar content in wine', x=0.5, y=1.02,fontsize=15)

```

```python
# Basic heat map to show correlation between features. 

plt.figure(figsize=(12,8))

data = red_wine_original.corr()

mask = np.triu(np.ones_like(data.corr()))

sns.heatmap(data, annot=True, cmap='Spectral', vmax=1, vmin=-1, mask = mask)

plt.show()
```

```python
# Basic heat map to show correlation between features. 

plt.figure(figsize=(12,8))

data = white_wine_original.corr()

mask = np.triu(np.ones_like(data.corr()))

sns.heatmap(data, annot=True, cmap='Spectral', vmax=1, vmin=-1, mask = mask)

plt.show()
```

```python
sns.jointplot(data=red_wine, x="quality", y="sulphates",color='red', marginal_kws=dict(bins=8, fill=True),).fig.suptitle("Title")
```

```python
sns.jointplot(data=white_wine, x="quality", y="sulphates",color='red', marginal_kws=dict(bins=10, fill=True),).fig.suptitle("Title")
```
