```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set()
```

```python
red_wine_original = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv',sep=';')
white_wine_original = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',sep=';')
#red_wine = red_wine_original.copy()
#white_wine = white_wine_original.copy()
```

```python
#red_wine.assign(wine_type='red')
#red_wine_original.assign(wine_type = lambda x:'red')
red_wine = red_wine_original.copy().assign(wine_type = lambda x:'red')

```

```python
red_wine
```

```python
white_wine = white_wine_original.copy().assign(wine_type = lambda x:'white')
```

```python
white_wine
```

```python
# Concatenating the 2 tables of Red and White wine

Frames = [red_wine, white_wine]
wines = pd.concat(Frames)

```

```python
wines
```

```python
wines_new=wines.drop_duplicates()
```
